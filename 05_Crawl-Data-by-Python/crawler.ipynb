1. Import selenium and interact with website on browser

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import pandas as pd
options = webdriver.ChromeOptions()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36")

driver = webdriver.Chrome(options=options)


a = 'https://www.chosic.com/music-genre-finder/'
driver.get(a) 
```

2. Navigating to song, artist,... drop down section

```python
drop_down = driver.find_element(By.NAME,'suggestion-options' )
select= Select(drop_down)
select.select_by_visible_text('Artist')
```

3. import Artist list from Excel file

```python
spotify_data = pd.read_csv('C:\\Users\\2001h\\documents\\Results.csv')
spotify_data.head()
```

4. extract Artist list from imported excel file

```python
artist_name = spotify_data['Artist'].tolist()
artist_name = list(filter(None,artist_name))
remove_nan = [element for element in artist_name if str(element) != "nan"]
artist_list = []
[artist_list.append(x) for x in remove_nan if x not in artist_list]
print(artist_list)
```

6. slide through artist_list and return value for genre

```python
from time import sleep
import numpy as np
search_field = driver.find_element(By.ID,'search-word')
x = []
cnt = 1
for i in artist_list:
    print(f'{cnt} - {i}')
    cnt = cnt + 1
    search_field.send_keys(i)
    sleep(2)
    try:
        search_button = driver.find_element(By.XPATH,'//*[@id="hh1"]')
        search_button.click()
        sleep(2)
        try:
            genre = driver.find_element(By.XPATH,'//*[@id="spotify-tags"]/div/div[2]/a[1]').text
            x.append(genre)
            print(genre)
        except:
            x.append(np.nan)
            print(genre)
    except:
        x.append(np.nan)
        print(genre)
    
    
    
    search_field.clear()
    sleep(0)
```
